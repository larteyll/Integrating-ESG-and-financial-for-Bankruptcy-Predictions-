```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set professional seaborn theme
sns.set_theme(style="whitegrid", palette="muted")

# Load dataset
df = pd.read_csv("sp500_esg_datafinal.csv")

# Display dataset overview
print("\nDataset Overview:")
print(df.head())

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing Values:")
print(missing_values)

# Handle missing values (numeric columns only)
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Check for duplicates
print("\nDuplicate Rows:", df.duplicated().sum())

# Correlation matrix
corr_matrix = df[numeric_cols].corr()

# ==========================================
# **🔹 Improved Data Visualizations**
# ==========================================

# 1️⃣ **Refined KDE Histograms (Elegant & Overlapping)**
plt.figure(figsize=(12, 6))
for col in numeric_cols[:4]:  # Limit to 4 for readability
    sns.kdeplot(df[col], fill=True, alpha=0.3, linewidth=2, label=col)

plt.title("Smoothed Distributions of Key Numerical Features", fontsize=16, fontweight='bold')
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()

# 2️⃣ **Boxen Plot (Better Than Boxplot for Large Data)**
plt.figure(figsize=(12, 6))
sns.boxenplot(data=df[numeric_cols[:5]], palette="coolwarm")
plt.title("Boxen Plot: Distribution of Key Features", fontsize=16, fontweight='bold')
plt.xlabel("Features")
plt.xticks(rotation=45)
plt.show()

# 3️⃣ **Bubble Chart (Polished & Emphasized)**
plt.figure(figsize=(10, 6))
x_col, y_col, size_col = numeric_cols[:3]  # Select 3 numeric variables
sns.scatterplot(
    data=df, x=x_col, y=y_col, size=size_col,
    sizes=(50, 500), alpha=0.7, edgecolor="black", linewidth=1.2,
    palette="coolwarm"
)

plt.title(f"Bubble Chart: {x_col} vs {y_col} (Bubble size = {size_col})", fontsize=16, fontweight='bold')
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.grid(alpha=0.3)
plt.show()

# 4️⃣ **Clean & Stylish Correlation Heatmap**
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="mako", linewidths=0.6, cbar=True, annot_kws={"size": 10})

plt.title("Refined Correlation Heatmap", fontsize=16, fontweight='bold')
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.show()




```
